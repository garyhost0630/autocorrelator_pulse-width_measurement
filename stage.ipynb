```python
from pipython import GCSDevice
import time

AXIS = "1"

POS_A = 6      # mm
POS_B = 6.5     # mm
VELOCITY = 0.3  # mm/s

LOOPS = 2        # <<< CHANGE THIS to desired number of back-and-forth cycles

POLL = 0.02
```

```python
pidevice = GCSDevice()
pidevice.InterfaceSetupDlg()   # Select USB C-884

print("IDN:", pidevice.qIDN())

# Turn servo on
pidevice.SVO(AXIS, 1)
print("Servo:", pidevice.qSVO(AXIS))
```

```python
def stage_pos():
    return float(pidevice.qPOS(AXIS)[AXIS])

def wait_on_target(timeout=60):
    t0 = time.time()
    while True:
        if bool(pidevice.qONT(AXIS)[AXIS]):
            return True
        if time.time() - t0 > timeout:
            return False
        time.sleep(POLL)

def move_abs(target):
    pidevice.MOV(AXIS, float(target))
    ok = wait_on_target()
    print(f"Reached {stage_pos():.6f} mm | ok={ok}")
```

```python
# Set velocity (mm/s)
pidevice.VEL(AXIS, VELOCITY)

try:
    current_vel = pidevice.qVEL(AXIS)[AXIS]
    print("Velocity set to:", current_vel, "mm/s")
except:
    print("Velocity set (qVEL not readable)")
```

```python
print("Moving to start position:", POS_A)
move_abs(POS_A)
```

```python
# ============================================
# Infinite Oscillation:
# Press "Interrupt Kernel" or Ctrl+C to stop
# ============================================

print("Starting infinite oscillation...")
print("Press Kernel Interrupt to stop.\n")

try:
    # Ensure we start at POS_A
    move_abs(POS_A)

    while True:
        move_abs(POS_B)
        move_abs(POS_A)

except KeyboardInterrupt:
    print("\nOscillation stopped by user.")
    print("Current position:", stage_pos())
```
