```python
import pandas as pd
import os

# Change this to your folder path
folder_path = your folder path here

data_list = []

# Loop from voltage_2.xlsx to voltage_11.xlsx
for i in range(1, 10):
    file_path = os.path.join(folder_path, f'voltage_{i}.xlsx')
    
    df = pd.read_excel(file_path, header=None)  # each file has only one column
    data_list.append(df)

# Concatenate vertically (stack them head-to-tail)
merged_df = pd.concat(data_list, ignore_index=True)

# Save the merged file
output_path = os.path.join(folder_path, 'voltage_all.xlsx')
merged_df.to_excel(output_path, index=False, header=False)

```
