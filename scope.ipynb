```python
import pyvisa
import pandas as pd
import os
import numpy as np

# connect
rm = pyvisa.ResourceManager()
scope = rm.open_resource("GPIB0::7::INSTR")

# setup
scope.timeout = 20000
scope.write(":TIMebase:MODE MAIN")
scope.write(":WAV:SOUR CHAN1")
scope.write(":WAV:FORM BYTE")
scope.write(":WAV:POIN 1000")

# get scaling info (only need Y values)
pre = list(map(float, scope.query(":WAV:PRE?").split(",")))
yinc, yorig, yref = pre[7], pre[8], int(pre[9])

# get RAW data
raw = scope.query_binary_values(":WAV:DATA?", datatype='B')

# convert to voltage
voltage = (np.array(raw) - yref) * yinc + yorig

# save
folder = r"C:\Users\NIL Lab\Desktop\test\data4"
os.makedirs(folder, exist_ok=True)

i = len(os.listdir(folder)) + 1
path = os.path.join(folder, f"voltage_{i-1}.xlsx")

pd.DataFrame(voltage).to_excel(path, index=False, header=False)

print("Saved:", path)
print("Mean voltage:", np.mean(voltage))
```
